```python
import pandas as pd
df = pd.read_csv("/content/Expt11.csv")
df
```

```python
# Count Students in Each Grade
# Find how many students obtained each grade
print("GRADE COUNT")
print(df['Grade'].value_counts())
print("GENDER COUNT")
print(df['Gender'].value_counts())
print("DEPARTMENT COUNT")
print(df['Department'].value_counts())
```

```python
# Percentage Distribution of Grades
print("Garde percentage")
print(df['Grade'].value_counts(normalize=True)*100)
```

```python
# Gender vs Grade Analysis
print("Gender vs Grade")
print(pd.crosstab(df['Gender'], df['Grade']))
```

```python
# Department vs Gender
print("Department vs Gender")
print(pd.crosstab(df['Department'], df['Gender']))
```

```python
# Department vs Grade
print("Department vs Grade")
print(pd.crosstab(df['Department'], df['Grade']))
```

```python
# Department vs Grade Percentage
print("Department vs Grade Percentage")
print(pd.crosstab(df['Department'], df['Grade'], normalize='index')*100)
```

```python
# Grouping data
print("Grouping data")
print(df.groupby('Department')['Grade'].value_counts())
```

```python
# ACTIVITY B
import pandas as pd

data = {'Order_ID':['O1','O2','O3','O4','O5','O6','O7','O8','O9','O10'],

'Category':['Electronics','Clothing','Electronics','Grocery','Clothing',
            'Electronics','Grocery','Clothing','Electronics','Grocery'],

'Payment_Method':['UPI','Card','UPI','Cash','Card',
                  'UPI','Cash','Card','UPI','Cash'],

'Delivery_Type':['Express','Standard','Express','Standard','Express',
                 'Standard','Express','Standard','Express','Standard'],

'Customer_Type':['New','Returning','Returning','New','Returning',
                 'New','Returning','New','Returning','New']}
df = pd.DataFrame(data)
df
```

```python
# Frequency Count
# Count how many orders ineach catergory
# Frequency of payment methods
display(df['Category'].value_counts())
display(df['Payment_Method'].value_counts())
```

```python
#Percentage Distribution
display(df['Category'].value_counts(normalize=True)*100)
```

```python
# Unique Values :  displays all unique categories
df['Category'].unique()
```

```python
# Number of Unique Values
df['Category'].nunique()
```

```python
# Cross Tabulation
# Category vs Payment
pd.crosstab(df['Category'], df['Payment_Method'])
```

```python
# Filtering Data
# Example : orders from electronics category

electronics_orders = df[df['Category'] == 'Electronics']
display(electronics_orders)
```

```python
# Sorting Data
sorted_df = df.sort_values(by='Order_ID')
display(sorted_df)
```

```python
# Grouping
# Orders by category and payment method
display(df.groupby(['Category', 'Payment_Method']).value_counts())
```
